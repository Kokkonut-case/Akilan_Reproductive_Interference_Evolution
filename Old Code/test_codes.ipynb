```python
import numpy as np
import matplotlib.pyplot as plt

def logistic_growth(N, R, K, generations):
    population = [N]
    for _ in range(generations - 1):
        N = (N * R * K) / (K + N * (R - 1))
        population.append(N)
    return population

# Define parameters
R = 1.5  # Growth rate
K = 100  # Carrying capacity
generations = 50  # Number of generations

# Define different starting values of N
initial_population_values = [100, 200, 300, 400]

# Plot population changes for each starting value of N
plt.figure(figsize=(10, 6))
for N in initial_population_values:
    population = logistic_growth(N, R, K, generations)
    plt.plot(range(1, generations + 1), population, label=f'Initial N={N}')

plt.title('Logistic Growth Simulation')
plt.xlabel('Generations')
plt.ylabel('Population')
plt.legend()
plt.grid(True)
plt.show()
```

```python
class PopulationSimulation:
    def __init__(self, popsize=2000, d=[0.2, 0.2], v=10, C=float('inf'), L=float('inf'),
                 mut=0.01, mutsize=0.1, initmean=[0, 0, 0], shared=False, generations=10001):
        self.popsize = popsize
        self.d = d
        self.v = v
        self.C = C
        self.L = L
        self.mut = mut
        self.mutsize = mutsize
        self.initmean = initmean                            #initialsing all parameters, including data sa, fa, etc
        self.shared = shared
        self.generations = generations
        self.warning = 0
        self.sex = 0
        self.sa = np.array([1, 2])
        self.ma = np.array([3, 4])
        self.fa = np.array([5, 6])
        self.trackmatings = min(20, self.popsize)
        self.data = np.full((int((generations - 1)/ 10 + 1), 4), np.nan)

    def initialize_population(self):
        self.pop = np.sign(np.random.randn(1, self.popsize))
        self.pop = np.vstack([self.pop, np.random.normal(0, 1, size=(6, self.popsize))])                    #making poulayi. init mean is more compact
        self.pop[self.sa[0]:self.fa[1], :] += self.initmean

    def run_simulation(self):
        for k in range(1, self.generations + 1):
            phenotype = self.calculate_phenotype()
            emergencetime = self.calculate_emergence_time(phenotype)
            deathtime = self.calculate_death_time(emergencetime)
            T_encounter, who_encounter = self.track_encounters(emergencetime, deathtime)                      #phenotype, calcula ohinel , emerence tue, t ejcger woencnter,                   
            relativefecundity, mateidentity = self.assign_mating(T_encounter, who_encounter, deathtime)
            self.update_data(k, phenotype, relativefecundity, mateidentity)
            self.produce_offspring(relativefecundity, mateidentity)
            self.mutate_population()

    def calculate_phenotype(self):                                                                            #mel and fel is defined

        mel = np.concatenate((self.sa, self.ma)) if self.shared else self.ma
        fel = np.concatenate((self.sa, self.fa)) if self.shared else self.fa                                    #relatobe fecnity is calculaed
        return (self.pop[self.sex, :] < 0) * np.mean(self.pop[mel, :], axis=0) + \
               (self.pop[self.sex, :] > 0) * np.mean(self.pop[fel, :], axis=0)

    def calculate_emergence_time(self, phenotype):                                                              #emeregnce time 
        return emergencesample(phenotype)                                                           

    def calculate_death_time(self, emergencetime):
        return emergencetime + ((self.pop[self.sex, :] < 0) * np.random.exponential(scale=1/self.d[0], size=self.popsize) +
                                (self.pop[self.sex, :] > 0) * np.random.exponential(scale=1/self.d[1], size=self.popsize))

    def track_encounters(self, emergencetime, deathtime):
        T_encounter = np.full((self.trackmatings, self.popsize), np.nan)                                    #track encounter T_Encoynert and who_encunter is made
        who_encounter = np.full((self.trackmatings, self.popsize), np.nan)
        for i in range(len(np.where(self.pop[self.sex, :] > 0)[0])):
            f = np.where(self.pop[self.sex, :] > 0)[0][i]
            t_encounter = emergencetime[f] + np.random.exponential(scale=1/self.v, size=self.popsize)
            t_encounter[self.pop[self.sex, :] > 0] = np.nan
            t_encounter[(t_encounter > deathtime[f]) & (self.pop[self.sex, :] > 0)] = np.nan
            sorted_indices = np.argsort(t_encounter)
            sorted_t_encounter = t_encounter[sorted_indices]
            sorted_whichmale = sorted_indices[:self.trackmatings]
            T_encounter[:self.trackmatings, f] = sorted_t_encounter[:self.trackmatings]
            who_encounter[:self.trackmatings, f] = sorted_whichmale[:self.trackmatings]
        return T_encounter, who_encounter

    def assign_mating(self, T_encounter, who_encounter, deathtime):
        matecount = np.zeros(self.popsize)
        mateidentity = np.full(self.popsize, np.nan)
        relativefecundity = np.zeros(self.popsize)
        while np.any(~np.isnan(T_encounter[0, :])):
            ind = np.nanargmin(T_encounter[0, :])
            time = T_encounter[0, ind]
            male = int(who_encounter[0, ind])
            if matecount[male] < self.C:
                T_encounter[:, ind] = np.nan
                matecount[male] += 1
                mateidentity[ind] = male
                relativefecundity[ind] = 400 * np.exp(time / self.L) / (1 + np.exp(time / self.L))**2
            else:
                T_encounter[:, ind] = np.concatenate((T_encounter[1:, ind], [np.nan]))
        if len(np.where(~np.isnan(T_encounter[-1, :]))[0]) > 0 and np.any(relativefecundity[np.where(~np.isnan(T_encounter[-1, :]))] == 0):
            self.warning = 1
        return relativefecundity, mateidentity

    def update_data(self, k, phenotype, relativefecundity, mateidentity):
        if (k - 1) % 10 == 0:
            mean_male_time = np.mean(np.mean(self.pop[self.ma, :][:, self.pop[self.sex, :] < 0], axis=0))
            mean_female_time = np.mean(np.mean(self.pop[self.fa, :][:, self.pop[self.sex, :] > 0], axis=0))
            mean_fecundity = np.mean(relativefecundity[~np.isnan(relativefecundity)])
            matelessness = 1 - (np.sum(~np.isnan(mateidentity)) / np.sum(self.pop[self.sex, :] > 0))
            self.data[int((k - 1) / 10), :] = [mean_male_time, mean_female_time, mean_fecundity, matelessness]
            if (k - 1) % 100 == 0:
                self.plot_data()

    def plot_data(self):
        plt.figure(1)
        t_for_axis = np.arange(0, self.generations, 10)
        plt.subplot(3, 1, 1)
        plt.plot(t_for_axis, self.data[:, 0], 'b', t_for_axis, self.data[:, 1], 'r')
        plt.ylabel('Timing')
        plt.title('Warning status: {}'.format(int(self.warning)))
        plt.subplot(3, 1, 2)
        plt.plot(t_for_axis, self.data[:, 2], 'g')
        plt.ylabel('Fecundity')
        plt.subplot(3, 1, 3)
        plt.plot(t_for_axis, self.data[:, 3], 'k')
        plt.ylabel('Matelessness')
        plt.tight_layout()
        plt.show()

    def produce_offspring(self, relativefecundity, mateidentity):
        fecundity_probability = relativefecundity / np.sum(relativefecundity)
        mother = np.random.choice(self.popsize, self.popsize, p=fecundity_probability, replace=True)
        sire = [int(mateidentity[x]) for x in mother]
        offspring = np.sign(np.random.randn(1, self.popsize))
        offspring = np.vstack([offspring, np.random.normal(0, 1, size=(6, self.popsize))])
        offspring[self.sa[0]:self.fa[1] + 1, :] = np.nan
        whichallele_egg = np.random.randint(2, size=(3, self.popsize))
        whichallele_sperm = np.random.randint(2, size=(3, self.popsize))
        for i in range(self.popsize):
            offspring[self.sa[0], i] = self.pop[self.sa[0] + whichallele_egg[0, i], mother[i]]
            offspring[self.ma[0], i] = self.pop[self.ma[0] + whichallele_egg[1, i], mother[i]]
            offspring[self.fa[0], i] = self.pop[self.fa[0] + whichallele_egg[2, i], mother[i]]
            offspring[self.sa[1], i] = self.pop[self.sa[0] + whichallele_sperm[0, i], sire[i]]
            offspring[self.ma[1], i] = self.pop[self.ma[0] + whichallele_sperm[1, i], sire[i]]
            offspring[self.fa[1], i] = self.pop[self.fa[0] + whichallele_sperm[2, i], sire[i]]
        self.pop[self.sa[0]:self.fa[1] + 1, :] = offspring[self.sa[0]:self.fa[1] + 1, :]

    def mutate_population(self):
        self.pop[self.sa[0]:self.fa[1] + 1, :] += (np.random.rand(6, self.popsize) < self.mut) * \
                                                  self.mutsize * np.random.randn(1, self.popsize)

sim = PopulationSimulation()
sim.initialize_population()
sim.run_simulation()


```

```python
import numpy as np
import matplotlib.pyplot as plt

class Parameters:   #class for defining parameters

    '''
Parameters defined as one class

A paramter instance is created for a specific paramter combination which is used by the simulation function'''



    def __init__(self, b1, b2, k1, k2, c12, c21, z_m1_initial, z_m2_initial, initial_population_species_1, initial_population_species_2, cost_to_males, strength_of_interference, V_A1, V_A2, num_generations):
        self.b1 = b1    #growth rate of sp. 1
        self.b2 = b2    #growth rate of sp. 2
        self.k1 = k1    #density dependent mortality of sp. 1
        self.k2 = k2    #density dependent mortality of sp. 2
        self.c12 = c12  #competetive effect of sp. 2 on sp. 1
        self.c21 = c21  #competetive effect of sp. 1 on sp. 2
        self.z_m1_initial = z_m1_initial  #initial mean discrimination of males of sp.1
        self.z_m2_initial = z_m2_initial  #initial mean discriminatiin of males of sp. 2
        self.initial_population_species_1 = initial_population_species_1  
        self.initial_population_species_2 = initial_population_species_2
        self.cost_to_males = cost_to_males   #cost of wrong matings for male fitness
        self.strength_of_interference = strength_of_interference  #how heterospecific matings affect fitness
        self.V_A1 = V_A1  #evolvabilty i.e. additive genetic variance of sp.1
        self.V_A2 = V_A2  #evolvabilty i.e. additive genetic variance of sp.2
        self.num_generations = num_generations  

class Simulation:

    '''Simulation defined as the other class which takes the parameter instance as an argument and runs the simulation. '''


    '''initial population of the species and initual trait values are recorded. Lists are created, which will store the population and trait values for each generation.'''
    def __init__(self, params):
        self.params = params
        self.population_species_history_1 = [params.initial_population_species_1]
        self.population_species_history_2 = [params.initial_population_species_2]
        self.z_m1_history = [params.z_m1_initial]
        self.z_m2_history = [params.z_m2_initial]
        self.z_m1 = params.z_m1_initial
        self.z_m2 = params.z_m2_initial


    def next_generation_ecology(self):
        '''function to calculate next generatioin pop numbers for both species using current population numbers and trait values. The new population numbers are calculated using the ecological equations given in the paper. The new population numbers are appended to the population history list.'''

    

        '''N1 and N2 are taken as the population of the species at the current generation.  (thus, last entry of the population history list)'''
        N1 = self.population_species_history_1[-1]
        N2 = self.population_species_history_2[-1]

        
        if N1 > 0:  #only if N1 > 0 is there a chance of there being population change. else the population continues to be 0
            next_N1 = self.params.b1 * (N1 / 2) * (N1 / (N1 + self.params.strength_of_interference * (1 - self.z_m2) * N2)) - self.params.k1 * N1 * (N1 + self.params.c12 * N2)
            next_N1 = max(next_N1, 0)   #to ensure that next_N1 is not negative 
        else:
            next_N1 = 0 
        
        if N2 > 0:
            next_N2 = self.params.b2 * (N2 / 2) * (N2 / (N2 + self.params.strength_of_interference * (1 - self.z_m1) * N1)) - self.params.k2 * N2 * (N2 + self.params.c21 * N1)
            next_N2 = max(next_N2, 0)   #to ensure that next_N2 is not negative
        else:
            next_N2 = 0

        self.population_species_history_1.append(next_N1)   #append next generation population numbers to the population history list
        self.population_species_history_2.append(next_N2)

    def next_generation_traits(self, N1, N2):
        if N1 > 0:  #only if both species have non zero population numbers, there is evolution of trait values. Else the trait values remain the same as the previous generation.
            W_m1 = 1 - (self.params.cost_to_males * (1 - self.z_m1) * (N2 / (N2 + N1)))   # ''Fitness'' of trait z_m1. fitness cost to mating, times proportion of incorrect matings
            
            dW_m1_dz_m2 = self.params.cost_to_males * (N2 / (N2 + N1)) #dw_m1/dz_m2 is the derivative of fitness with respect to the trait value of the species.
            
            delta_z_m1 = self.params.V_A1 * (1 / W_m1) * dW_m1_dz_m2  #change in trait value of species 1. Additive genetic variance x 1/W x w
            new_z_m1 = self.z_m1 + delta_z_m1
            
            if new_z_m1 <= 1:
                self.z_m1 = new_z_m1

            self.z_m1_history.append(self.z_m1)

        if N2 > 0: 

            W_m2 = 1 - (self.params.cost_to_males * (1 - self.z_m2) * (N2 / (N2 + N1)))
            dW_m2_dz_m2 = self.params.cost_to_males * (N1 / (N1 + N2))
            delta_z_m2 = self.params.V_A2 * (1 / W_m2) * dW_m2_dz_m2
            new_z_m2 = self.z_m2 + delta_z_m2
            if new_z_m2 <= 1:
                self.z_m2 = new_z_m2

            self.z_m2_history.append(self.z_m2)

    def next_generation(self):
        N1 = self.population_species_history_1[-1]  #from the list take the last entry which is the population number for the next generation
        N2 = self.population_species_history_2[-1] 
        self.next_generation_traits(N1, N2)  #using the new population numbers can 
        self.next_generation_ecology()   #calculate and update the population numbers for the next generation
    
        return N1, N2

    def run_simulation(self):
        for t in range(self.params.num_generations - 1):
            N1, N2 = self.next_generation()
            if N1 == 0 or N2 == 0:
                print(f"At generation {t + 1}, one of the species went extinct.")
                break
        return self.population_species_history_1, self.population_species_history_2, self.z_m1_history, self.z_m2_history

class Plotter:
    @staticmethod
    def plot_population_dynamics(population_species_1, population_species_2):
        plt.ion()  # Turn on interactive mode
        fig, ax = plt.subplots(figsize=(10, 5))
        ax.plot(population_species_1, label='Species 1')
        ax.plot(population_species_2, label='Species 2')
        ax.set_xlabel('Time')
        ax.set_ylabel('Population')
        ax.set_title('Population Dynamics of Two Competing Species')
        ax.legend()
        ax.grid(True)
        plt.show()

    @staticmethod
    def plot_traits_evolution(z_m1_history, z_m2_history):
        plt.ion()  # Turn on interactive mode
        fig, ax = plt.subplots(figsize=(10, 5))
        ax.plot(z_m1_history, label="Mean Trait z_m1")
        ax.plot(z_m2_history, label="Mean Trait z_m2")
        ax.set_xlabel("Generation")
        ax.set_ylabel("Mean Trait Value")
        ax.set_title('Evolution of Traits')
        ax.legend()
        ax.grid(True)
        plt.show()

# Parameters
params = Parameters(
    b1 = 2.5, 
    b2 = 2.5, 
    k1 = 0.00001, 
    k2 = 0.00001, 
    c12 = 0.1, 
    c21 = 0.1, 
    z_m1_initial = 0.5, 
    z_m2_initial = 0.5, 
    initial_population_species_1 = 10, 
    initial_population_species_2 = 10, 
    cost_to_males = 0.1,
    strength_of_interference = 0.1,
    V_A1 = 0.01, 
    V_A2 = 0.1, 
    num_generations = 1000
)

# Simulation
sim = Simulation(params)
population_species_1, population_species_2, z_m1_history, z_m2_history = sim.run_simulation()

# Plotting
Plotter.plot_population_dynamics(population_species_1, population_species_2)
Plotter.plot_traits_evolution(z_m1_history, z_m2_history)

print("Final mean trait z_m1:", sim.z_m1)
print("Final mean trait z_m2:", sim.z_m2)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import plotly.graph_objs as go
import plotly.io as pio

class Parameters:
    def __init__(self, b1, b2, k1, k2, c12, c21, z_m1_initial, z_m2_initial, initial_population_species_1, initial_population_species_2, cost_to_males, strength_of_interference, V_A1, V_A2, num_generations):
        self.b1 = b1
        self.b2 = b2
        self.k1 = k1
        self.k2 = k2
        self.c12 = c12
        self.c21 = c21
        self.z_m1_initial = z_m1_initial
        self.z_m2_initial = z_m2_initial
        self.initial_population_species_1 = initial_population_species_1
        self.initial_population_species_2 = initial_population_species_2
        self.cost_to_males = cost_to_males
        self.strength_of_interference = strength_of_interference
        self.V_A1 = V_A1
        self.V_A2 = V_A2
        self.num_generations = num_generations

class Simulation:
    def __init__(self, params):
        self.params = params
        self.population_species_history_1 = [params.initial_population_species_1]
        self.population_species_history_2 = [params.initial_population_species_2]
        self.z_m1_history = [params.z_m1_initial]
        self.z_m2_history = [params.z_m2_initial]
        self.z_m1 = params.z_m1_initial
        self.z_m2 = params.z_m2_initial

    def next_generation_ecology(self):
        N1 = self.population_species_history_1[-1]
        N2 = self.population_species_history_2[-1]
        
        if N1 > 0.01:
            next_N1 = self.params.b1 * (N1 / 2) * (N1 / (N1 + self.params.strength_of_interference * (1 - self.z_m2) * N2)) - (self.params.k1 * N1 * (N1 + self.params.c12 * N2))
            next_N1 = max(next_N1, 0)
        else:
            next_N1 = 0 
        
        if N2 > 0.01:
            next_N2 = self.params.b2 * (N2 / 2) * (N2 / (N2 + self.params.strength_of_interference * (1 - self.z_m1) * N1)) - (self.params.k2 * N2 * (N2 + self.params.c21 * N1))
            next_N2 = max(next_N2, 0)
        else:
            next_N2 = 0

        self.population_species_history_1.append(next_N1)
        self.population_species_history_2.append(next_N2)

    def next_generation_traits(self, N1, N2):
        if N1 > 0.01:
            W_m1 = 1 - (self.params.cost_to_males * (1 - self.z_m1) * (N2 / (N2 + N1)))
            dW_m1_dz_m2 = self.params.cost_to_males * (N2 / (N2 + N1))
            delta_z_m1 = self.params.V_A1 * (1 / W_m1) * dW_m1_dz_m2
            new_z_m1 = self.z_m1 + delta_z_m1
            
            if new_z_m1 <= 1:
                self.z_m1 = new_z_m1
            else :
                self.z_m1 = 1

            self.z_m1_history.append(self.z_m1)

        if N2 > 0.01:
            W_m2 = 1 - (self.params.cost_to_males * (1 - self.z_m2) * (N2 / (N2 + N1)))
            dW_m2_dz_m2 = self.params.cost_to_males * (N1 / (N1 + N2))
            delta_z_m2 = self.params.V_A2 * (1 / W_m2) * dW_m2_dz_m2
            new_z_m2 = self.z_m2 + delta_z_m2
            
            if new_z_m2 <= 1:
                self.z_m2 = new_z_m2
            else :
                self.z_m2 = 1

            self.z_m2_history.append(self.z_m2)

    def next_generation(self):
        self.next_generation_ecology()
        N1 = self.population_species_history_1[-1]
        N2 = self.population_species_history_2[-1]
        self.next_generation_traits(N1, N2)
        return N1, N2

    def run_simulation(self):
        for t in range(self.params.num_generations - 1):
            N1, N2 = self.next_generation()
            if N1 < 0.01 or N2 < 0.01:
                break

        return self.population_species_history_1, self.population_species_history_2, self.z_m1_history, self.z_m2_history

class outcome_stats_and_plots:
    def __init__(self, population_species_1, population_species_2, z_m1_history, z_m2_history, params):
        self.population_species_1 = population_species_1
        self.population_species_2 = population_species_2
        self.z_m1_history = z_m1_history
        self.z_m2_history = z_m2_history
        self.params = params

        self.outcome = None
        self.number_of_generations_until_extinctioin = None
        self.final_population_species_1 = None
        self.final_population_species_2 = None
        self.final_z_m1 = None
        self.final_z_m2 = None

        self.outcome_stats()
        
    def plot_population_dynamics_and_trait_evolution(self):
        self.plot_population_dynamics()
        self.plot_traits_evolution()


    def plot_population_dynamics(self):
        fig = go.Figure()

        # Population Species 1
        fig.add_trace(go.Scatter(x=list(range(len(self.population_species_1))), 
                                 y=self.population_species_1, 
                                 mode='lines', 
                                 name='Species 1 (Slow-ev)', 
                                 line=dict(color='blue')))

        # Population Species 2
        fig.add_trace(go.Scatter(x=list(range(len(self.population_species_2))), 
                                 y=self.population_species_2, 
                                 mode='lines', 
                                 name='Species 2 (Fast-ev)', 
                                 line=dict(color='red')))

        fig.update_layout(
            title='Population Dynamics of Two Competing Species',
            xaxis_title='Time',
            yaxis_title='Population',
            legend_title='Species',
            template='plotly_white'
        )

        pio.show(fig)

    def plot_traits_evolution(self):
        fig = go.Figure()

        # Mean Trait z_m1
        fig.add_trace(go.Scatter(x=list(range(len(self.z_m1_history))), 
                                 y=self.z_m1_history, 
                                 mode='lines', 
                                 name='Mean Trait z_m1', 
                                 line=dict(color='blue')))

        # Mean Trait z_m2
        fig.add_trace(go.Scatter(x=list(range(len(self.z_m2_history))), 
                                 y=self.z_m2_history, 
                                 mode='lines', 
                                 name='Mean Trait z_m2', 
                                 line=dict(color='red')))

        fig.update_layout(
            title='Evolution of Traits',
            xaxis_title='Generation',
            yaxis_title='Mean Trait Value',
            legend_title='Traits',
            template='plotly_white'
        )

        pio.show(fig)

    def outcome_stats(self):

        if self.population_species_1[-1] < 0.01 and self.population_species_2[-1] < 0.01:
            self.outcome = "both extinct"
        elif self.population_species_1[-1] < 0.01 and self.population_species_2[-1] > 0.01:
            self.outcome = "sp 2 (fast-ev) wins, sp 1 (slow-ev) extinct"
        elif self.population_species_2[-1] < 0.01 and self.population_species_1[-1] > 0.01:
            self.outcome = "sp 1 (slow-ev) wins, sp 2 (fast-ev) extinct"
        else:
            self.outcome = "both coexist"

        if len(self.population_species_1) < self.params.num_generations:
            self.number_of_generations_until_extinctioin = len(self.population_species_1)
            self.final_population_species_1 = self.population_species_1[-1]
            self.final_population_species_2 = self.population_species_2[-1]
            self.final_z_m1 = self.z_m1_history[-1]
            self.final_z_m2 = self.z_m2_history[-1]

    def print_outcome_stats(self):

        print("Outcome of simulation: ", self.outcome)    
        print("Number of generations until extinction: ", self.number_of_generations_until_extinctioin)
        print("Final population of species 1: ", self.final_population_species_1)
        print("Final population of species 2: ", self.final_population_species_2)
        print("Final trait value z_m1: ", self.final_z_m1)
        print("Final trait value z_m2: ", self.final_z_m2)

def run_simulation_and_plot_pop_dynamics_and_trait_evolution(outcome_stats_and_plots_instance):
    outcome_stats_and_plots_instance.plot_population_dynamics_and_trait_evolution()
    outcome_stats_and_plots_instance.print_outcome_stats()

def run_simulation_and_get_outcome_instance(params):
    simulation = Simulation(params)
    population_species_1, population_species_2, z_m1_history, z_m2_history = simulation.run_simulation()
    outcome_stats_and_plots_instance = outcome_stats_and_plots(population_species_1, population_species_2, z_m1_history, z_m2_history, params)
    
    return outcome_stats_and_plots_instance


```

```python
import plotly.graph_objects as go
import plotly.express as px
import numpy as np

class EvolutionSimulation:
    def __init__(self, parameters):
        self.parameters = parameters
        self.z_m1 = 0
        self.z_m2 = 0
        self.z_m1_history = []
        self.z_m2_history = []
    
    def next_generation_traits(self, N1, N2):
        if N1 > 0.01:
            delta_z_m1 = self.parameters.V_A1 * (self.parameters.conspecfic_mating_rate_multiplier_sp_1) / (1 + self.parameters.conspecfic_mating_rate_multiplier_sp_1*self.z_m1)
            new_z_m1 = self.z_m1 + delta_z_m1
            
            if new_z_m1 <= 1:
                self.z_m1 = new_z_m1
            else:
                self.z_m1 = 1

            self.z_m1_history.append(self.z_m1)



        if N2 > 0.01:
            W_m2 = 1 - self.parameters.conspecfic_mating_rate_multiplier_sp_2*(((1 - self.z_m2) * (N1 / (N1 + N2))) ** (1/self.parameters.conspecfic_mating_rate_multiplier_sp_2))
            dW_m2_dz_m2 = ((1 - self.z_m2) ** (1/self.parameters.conspecfic_mating_rate_multiplier_sp_2 - 1)) * (N1 / (N1 + N2)) ** ((1/self.parameters.conspecfic_mating_rate_multiplier_sp_2))
            delta_z_m2 = self.parameters.V_A2 * (1 / W_m2) * dW_m2_dz_m2
            new_z_m2 = self.z_m2 + delta_z_m2
            
            if new_z_m2 <= 1:
                self.z_m2 = new_z_m2
            else:
                self.z_m2 = 1

            self.z_m2_history.append(self.z_m2)

    def run_simulation(self, N1, N2, generations):
        for _ in range(generations):
            self.next_generation_traits(N1, N2)
            
# Example parameters object (to be defined as needed)
class Parameters:
    def __init__(self, conspecfic_mating_rate_multiplier_sp_1, conspecfic_mating_rate_multiplier_sp_2, V_A1, V_A2):
        self.conspecfic_mating_rate_multiplier_sp_1 = conspecfic_mating_rate_multiplier_sp_1
        self.conspecfic_mating_rate_multiplier_sp_2 = conspecfic_mating_rate_multiplier_sp_2
        self.V_A1 = V_A1
        self.V_A2 = V_A2

# Running simulations for different values of a and plotting
a_values = np.linspace(0.1, 0.8, 5)
N1, N2 = 100, 100
generations = 1000

fig_z_vs_gen = go.Figure()
fig_delta_z_vs_z = go.Figure()


for a in a_values:
    params = Parameters(conspecfic_mating_rate_multiplier_sp_1=a,conspecfic_mating_rate_multiplier_sp_2 = a, V_A1=0.01, V_A2=0.01)
    sim = EvolutionSimulation(params)
    sim.run_simulation(N1, N2, generations)
    
    generations_range = list(range(generations))
    
    fig_z_vs_gen.add_trace(go.Scatter(x=generations_range, y=sim.z_m1_history, mode='lines', name=f'z_m1, a={a}'))
    # fig_z_vs_gen.add_trace(go.Scatter(x=generations_range, y=sim.z_m2_history, mode='lines', name=f'z_m2, a={a}'))
    
    delta_z_m1 = [j-i for i, j in zip(sim.z_m1_history[:-1], sim.z_m1_history[1:])]
    fig_delta_z_vs_z.add_trace(go.Scatter(x=sim.z_m1_history[:-1], y=delta_z_m1, mode='lines+markers', name=f'Delta z_m1, a={a}'))

    # delta_z_m2 = [j-i for i, j in zip(sim.z_m2_history[:-1], sim.z_m2_history[1:])]
    # fig_delta_z_vs_z.add_trace(go.Scatter(x=sim.z_m2_history[:-1], y=delta_z_m2, mode='lines+markers', name=f'Delta z_m2, a={a}'))

fig_z_vs_gen.update_layout(title='z vs Generation', xaxis_title='Generation', yaxis_title='z')
fig_delta_z_vs_z.update_layout(title='z vs z', xaxis_title='z', yaxis_title='z')

fig_z_vs_gen.show()
fig_delta_z_vs_z.show()
```

```python
import numpy as np
import plotly.graph_objects as go

# Parameters for the Gaussian distribution
mu = 0.5       # Mean
sigma = 0.1    # Standard deviation

# Generate sample data from a Gaussian distribution
np.random.seed(0)  # For reproducibility
sample_data = np.random.normal(mu, sigma, 10000)

# Calculate histogram
num_bins = 1000  # Large number of bins for fine resolution
hist_data, bins = np.histogram(sample_data, bins=num_bins, range=[0, 1], density=True)

# Calculate bin centers
bin_centers = (bins[:-1] + bins[1:]) / 2

# Calculate Gaussian density function
z = np.linspace(0, 1, 1000)
f_z = (1 / (sigma * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((z - mu) / sigma) ** 2)

# Create the plot
fig = go.Figure()

# Add histogram trace
fig.add_trace(go.Histogram(
    x=sample_data,
    histnorm='probability density',
    name='Histogram',
    nbinsx=num_bins,
    xbins=dict(start=0, end=1),
    showlegend=False  # Hide histogram from the legend
))

# Add Gaussian density plot trace
fig.add_trace(go.Scatter(
    x=z,
    y=f_z,
    mode='lines',
    name='Gaussian Density',
    line=dict(color='red'),
    showlegend=False  # Hide density plot from the legend
))

# Add vertical line for mean
fig.add_trace(go.Scatter(
    x=[mu, mu],
    y=[0, max(f_z)],
    mode='lines',
    line=dict(color='blue', dash='dash'),
    name='Mean (\u03BC)'  # Label for the mean line
))

# Update layout
fig.update_layout(
    title='Normalized Histogram with Gaussian Density Plot',
    xaxis_title='z (Discrimination)',
    yaxis_title='Density',
    xaxis=dict(range=[0, 1]),
    yaxis=dict(title='Density')
)

# Show plot
fig.show()
```
